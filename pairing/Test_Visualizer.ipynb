```python
!pip install seaborn
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter
from pathlib import Path
import pickle
import itertools
import seaborn as sns
import os

import time
```

```python
def plotting(x,y):
    sns.barplot(x=y,y=x,orient='h')
```

```python
def read_prev_grps():

    try: 
        with open('prev_pairings.pickle', 'rb') as handle:
            prev_pairs = pickle.load(handle)

        #print('These are the groups from the past: {}'.format(prev_groups))
    except: 
        'CANNOT READ PREVIOUS GROUPS.'
        prev_pairs = []
    return prev_pairs
```

```python
lst = read_prev_grps()

lst = list(itertools.chain(*lst))
lst = [tuple(sorted(pair)) for pair in lst] 
res = Counter(lst)
print(res)
```

```python
y = dict(res)
plt.rcParams["figure.figsize"] = (3,15)
counts = list(y.values())
x = np.linspace(0,1,len(counts))
print(len(x))
print(len(counts))
fig, ax = plt.subplots()
plotting(x,counts)
ax.set_yticklabels(y.keys())
ax.set_xlabel('frequency of group')
plt.show
```
